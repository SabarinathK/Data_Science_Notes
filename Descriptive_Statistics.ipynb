```python
X['sex']=X['sex'].replace({'Female':0, 'Male':1})
X['smoker']=X['smoker'].replace({'No':0, 'Yes':1})
X['day']=X['day'].replace({'Sun':0, 'Sat':1, 'Thur':3, 'Fri':4})
X['time']=X['time'].replace({'Dinner':0, 'Lunch':1})
```
